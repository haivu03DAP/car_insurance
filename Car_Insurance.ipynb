## Feature Engineering

There are three types of features:
* Client features:-  Age, Job, Marital, Education, Default, Balance, HHInsurance, CarLoan
* Communication features:-  LastContactDay, LastContactMonth, CallStart, CallEnd, Communication, NoOfContacts, DaysPassed
* Previous campaign features:-  PrevAttempts, Outcome

```python
# First simplify some client features
# Create age group based on age
df['AgeBand']=pd.cut(df['Age'],5)
print(df['AgeBand'].value_counts())
```

```python
# create the age groups based on AgeBand
df.loc[(df['Age']>=17) & (df['Age']<34),'AgeBin'] = 1
df.loc[(df['Age']>=34) & (df['Age']<49),'AgeBin'] = 2
df.loc[(df['Age']>=49) & (df['Age']<65),'AgeBin'] = 3
df.loc[(df['Age']>=65) & (df['Age']<80),'AgeBin'] = 4
df.loc[(df['Age']>=80) & (df['Age']<96),'AgeBin'] = 5
# changing datatype to int
df['AgeBin'] = df['AgeBin'].astype(int)
```

```python
#create balance group balance
df['BalanceBand']=pd.cut(df['Balance'],5)
print(df['BalanceBand'].value_counts())
```

```python
# creating age groups based on Balancebinned
df.loc[(df['Balance']>=-3200) & (df['Balance']<17237),'BalanceBin'] = 1
df.loc[(df['Balance']>=17237) & (df['Balance']<37532),'BalanceBin'] = 2
df.loc[(df['Balance']>=37532) & (df['Balance']<57827),'BalanceBin'] = 3
df.loc[(df['Balance']>=57827) & (df['Balance']<78122),'BalanceBin'] = 4
df.loc[(df['Balance']>=78122) & (df['Balance']<98418),'BalanceBin'] = 5
# converting the datatype
df['BalanceBin'] = df['BalanceBin'].astype(int)
```

There seems to be a unique problem with respect to the CallStart and CallEnd attributes and are recorded as object variables which can be computed easily using the datetime function, so converting it to datetime function and subtracting them arrives at the actual CallTime which can be further binned as above.

```python
#Converting CallStart and CallEnd to datetime datatype
df['CallStart'] = pd.to_datetime(df['CallStart'] )
df['CallEnd'] = pd.to_datetime(df['CallEnd'] )
#Subtracting both the Start and End times to arrive at the actual CallTime
df['CallTime'] = (df['CallEnd'] - df['CallStart']).dt.total_seconds()
```

```python
#Binning the CallTime
df['CallLenBand']=pd.cut(df['CallTime'],5)
print(df['CallLenBand'].value_counts())
```

```python
# Create call length bins
df.loc[(df['CallTime']>= 0) & (df['CallTime']<650),'CallTimeBin'] = 1
df.loc[(df['CallTime']>=650) & (df['CallTime']<1300),'CallTimeBin'] = 2
df.loc[(df['CallTime']>=1300) & (df['CallTime']<1900),'CallTimeBin'] = 3
df.loc[(df['CallTime']>=1900) & (df['CallTime']<2600),'CallTimeBin'] = 4
df.loc[(df['CallTime']>=2600) & (df['CallTime']<3300),'CallTimeBin'] = 5
# converting the datatype
df['CallTimeBin'] = df['CallTimeBin'].astype(int)
df = df.drop('CallLenBand',axis=1)
```

```python
# Repalceing the month with their no value
df['LastContactMonth']=df['LastContactMonth'].replace(['jan', 'may', 'jun', 'mar', 'nov', 'jul', 'aug', 'sep', 'apr',
       'feb', 'oct', 'dec'],[1,5,6,3,11,7,8,9,4,2,10,12])
#converting datatype
df['LastContactMonth']=df['LastContactMonth'].astype(int)
```

```python
# Convert education level to numeric
df['Education'] = df['Education'].replace({'None':0,'primary':1,'secondary':2,'tertiary':3})
```

```python
#Dropping the original columns of the binned, just to make things easy
df.drop(['Id','Age','Balance','CallStart','CallEnd','BalanceBand','AgeBand'],axis = 1,inplace = True)
```

```python
df.info()
```

## Label Encoding

```python
## differentiate by types
df_numeric=df.select_dtypes(include=np.number)

df_cat=df.select_dtypes(include='object')
```

```python
## getting dummies values for the categorical ariables i.e one hot encoding
df_cat_dummies=pd.get_dummies(df_cat)
```

```python
## gathering final data after getting dummy values
df_final=pd.concat((df_numeric,df_cat_dummies),axis=1)
```

```python
df_final.info()
```

## Building the Model(s)

```python
# Dropping the Target for X
x= df_final.drop(['CarInsurance'],axis=1)
# Including only the Target for y
y=df_final['CarInsurance']
```

```python
#Splitting the Training and Testing data having 30% of Test data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30,random_state=110)
```

```python
#Getting the shape of new test and train data
print('x_test data shape :-',x_test.shape)
print('x_train data shape :-',x_train.shape)
print('y_test data shape :-',y_test.shape)
print('y_train data shape :-',y_train.shape)
```
